```python
def get_mean_and_SE(percentages):

    means = np.mean(percentages, axis=0)
    
    SE = scipy.stats.sem(percentages, axis=0)

    return means, SE
```

```python
# Specify the path to your CSV file
csv_file_path = "L:/widefield/max_percent_day_1_saline.csv"

# Load the CSV file into a NumPy array
max_percent_day_1_saline = np.loadtxt(csv_file_path, delimiter=',')
```

```python
# Specify the path to your CSV file
csv_file_path = "L:/widefield/max_percent_day_1_psilocybin.csv"

# Load the CSV file into a NumPy array
max_percent_day_1_psilocybin = np.loadtxt(csv_file_path, delimiter=',')
```

```python
max_percent_day_1_saline_means, max_percent_day_1_saline_SE = get_mean_and_SE(max_percent_day_1_saline)
max_percent_day_1_psilocybin_means, max_percent_day_1_psilocybin_SE = get_mean_and_SE(max_percent_day_1_psilocybin)
```

```python
import scipy.stats as stats

data = max_percent_diffs_day_1_6_psilocybin[:,0]

# Conducting the one-sample Wilcoxon signed-rank test
statistic, p_value = stats.wilcoxon(data - 0, alternative='less')

alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The sample median is significantly less than zero.")
else:
    print("Fail to reject the null hypothesis: The sample median is not significantly less than zero.")

print(p_value, statistic)
```

```python
data = max_percent_diffs_day_1_6_saline[:,6]

# Conducting the one-sample t-test
t_statistic, p_value = scipy.stats.ttest_1samp(data, 0)

# Since we want a one-tailed test, we halve the p-value
p_value_one_tailed = p_value / 2

# For a left-tailed test, check if the t-statistic is negative and the p-value is less than alpha
alpha = 0.05
if t_statistic < 0 and p_value_one_tailed < alpha:
    print("Reject the null hypothesis: The sample mean is significantly less than zero.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly less than zero.")

print(p_value_one_tailed, t_statistic)
```
